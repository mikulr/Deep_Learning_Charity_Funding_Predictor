## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.decomposition import PCA
#  Import and read the charity_data.csv.
url= "https://raw.githubusercontent.com/mikulr/deep_learning/main/Resources/charity_data.csv"
application_df = pd.read_csv(url)
application_df.head()
```

First Red Flag= "Special Considerations"

```python
specials = application_df[application_df.SPECIAL_CONSIDERATIONS == 'Y']
specials
# no easily apparent patterns, but as name says it already has something different about it so lets drop it just to be safe
```

```python
application_df = application_df[application_df.SPECIAL_CONSIDERATIONS == 'N']
application_df
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME' and STATUS and 'SPECIAL_CONSIDERATIONS'
application_df=application_df.drop(columns={'EIN','NAME', "STATUS", 'SPECIAL_CONSIDERATIONS'})
```

```python
# Determine the number of unique values in each column.
counts = application_df.nunique()
counts
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_count= application_df['APPLICATION_TYPE'].value_counts()

app_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_count[app_count <15].index)
# application_types_to_replace
```

```python
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_count= application_df['CLASSIFICATION'].value_counts()
class_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_count2=application_df['CLASSIFICATION'].value_counts().loc[lambda x : x>1]
# class_count2
```

```python
classifications_to_replace = list(class_count[class_count < 50].index)
# classifications_to_replace
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
recounts = application_df.nunique()
recounts
```

```python
# Create Filter for outliers in the ASK_AMT column
q = application_df['ASK_AMT'].quantile(0.90)

no_out= application_df[application_df['ASK_AMT'] < q]
```

```python
recounts2 = no_out.nunique()
recounts2
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_dummies= pd.get_dummies(no_out)
app_dummies
```

```python
# Split our preprocessed data into our features and target arrays
X = app_dummies.drop(["IS_SUCCESSFUL"], axis="columns")
y = app_dummies["IS_SUCCESSFUL"]

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn_model = tf.keras.models.Sequential()

# Input layer
nn_model.add(tf.keras.layers.Dense(units=70, activation="tanh", input_dim=54))

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=50, activation="tanh"))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=50, activation="tanh"))

# Third hidden layer
nn_model.add(tf.keras.layers.Dense(units=50, activation="tanh"))

# Fourth hidden layer
nn_model.add(tf.keras.layers.Dense(units=25, activation="tanh"))

# Fifth hidden layer
nn_model.add(tf.keras.layers.Dense(units=25, activation="tanh"))

# Sixth hidden layer
nn_model.add(tf.keras.layers.Dense(units=25, activation="tanh"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh'])
    
    # Allow kerastuner to decide number of neurons in first layer
    nn.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=25,
        max_value=55,
        step=10), activation=activation, input_dim=54))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 12)):
        nn.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=71,
            step=10),
            activation=activation))
    
    nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    
    return nn
```

```python
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=50,
    hyperband_iterations=2, 
    overwrite=True)
```

```python
tuner.search(X_train_scaled,y_train,validation_data=(X_test_scaled,y_test), callbacks=[tf.keras.callbacks.EarlyStopping("val_accuracy", patience=3)])
```

```python
best_hyper = tuner.get_best_hyperparameters(1)[0]
best_hyper.values
```

```python
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
